# 전처리 및 모델링 적용 후 시각화

```python
# SSG 랜더스 댓글 감정 석
# 1. 감정 분석 파이프라인 초기화
pipe = pipeline(
    "zero-shot-classification",
    model="joeddav/xlm-roberta-large-xnli",
    device=0
)
candidate_labels = ["positive", "negative"] # 긍정, 부정

# 2. 데이터 로딩
df_ssg = pd.read_csv("ssg_comment.csv")
df_ssg["comment"] = df_ssg["comment"].astype(str).str.strip() # 문자열로 변환 및 공백 제거

# 3. 전처리
def clean_comment(text):
    text = re.sub(r"[^가-힣a-zA-Z0-9\s]", "", text) # 한글, 영문, 숫자, 공백 제외 제거
    text = re.sub(r"\s+", " ", text).strip() # 중복 공백 제거
    return text

df_ssg["cleaned_comment"] = df_ssg["comment"].apply(clean_comment)
df_ssg = df_ssg[df_ssg["cleaned_comment"].str.len() > 5]            # 짧은 댓글 제거
df_ssg = df_ssg.drop_duplicates(subset=["cleaned_comment"])       # 중복 제거
df_ssg = df_ssg.dropna(subset=["cleaned_comment"]).copy()         # 결측치 제거

# 4. 감정 분석 실행
results = [pipe(text, candidate_labels) for text in df_ssg["cleaned_comment"]]
df_ssg["label"] = [r["labels"][0] for r in results]
df_ssg["score"] = [r["scores"][0] for r in results]

# 5. 감정 비율 계산
sentiment_counts = df_ssg["label"].value_counts(normalize=True) * 100

# 6. 시각화
colors = {"positive": "skyblue", "negative": "red"}
pie_colors = [colors[label] for label in sentiment_counts.index]
plt.figure(figsize=(6, 6))
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct="%1.1f%%",
    colors=pie_colors,
    startangle=140
)
plt.title("SSG랜더스 댓글 감정 분포 (긍정 vs 부정)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 1. 감정 분석 파이프라인 초기화
pipe = pipeline(
    "zero-shot-classification",
    model="joeddav/xlm-roberta-large-xnli",
    device=0  # GPU 사용
)
candidate_labels = ["positive", "negative"]

# 2. 데이터 로딩
df_lotte = pd.read_csv("lotte_comment.csv")

# 3. 전처리
if "comment" not in df_lotte.columns:
    raise ValueError("'comment' 컬럼이 존재하지 않습니다.")

# 문자열로 변환 및 공백 제거
df_lotte["comment"] = df_lotte["comment"].astype(str).str.strip()

# 특수문자 및 이모지 제거 함수
def clean_comment(text):
    text = re.sub(r"[^가-힣a-zA-Z0-9\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# 공백/결측치/중복/짧은 댓글 제거
df_lotte["cleaned_comment"] = df_lotte["comment"].apply(clean_comment)
df_lotte = df_lotte[df_lotte["cleaned_comment"].str.len() > 5]
df_lotte = df_lotte.drop_duplicates(subset=["cleaned_comment"])
df_lotte = df_lotte.dropna(subset=["cleaned_comment"]).copy()

# 4. 감정 분석 실행
results = [pipe(text, candidate_labels) for text in df_lotte["cleaned_comment"]]
df_lotte["label"] = [r["labels"][0] for r in results]
df_lotte["score"] = [r["scores"][0] for r in results]

# 5. 키워드 기반 후처리
# 긍/부정 키워드 생성
positive_keywords = [
    "울컥", "지린다", "미쳐", "미쳤다", "소름", "분위기 달랐다", "외쳐", "미라클",
    "껴안고", "맘 놓고 본다", "절대 지켜", "살아난", "완치", "스트라이크", "끝내기",
    "야잘잘", "그 잡채", "파이팅", "고생하셨습니다", "고생 많다",
    "진짜 말 안된다", "수비 폼 미쳤다", "전캡틴", "정정", "정답은 롯데드라",
    "롯며들어", "사랑", "믿고", "보물발견", "공감되면 개추", "완전 반했어",
    "어떻하니", "선수층 말이 안된다", "피해갈 타순이 없다", "화이팅"
    "자티비", "전구단", "마음속", "아싸", "욕봤다", "연봉치 다 받았다"
]

negative_keywords = [
    "빠따들 엎드려", "안터지", "내다버린", "2군", "염색 금지령", "노니",
    "가지가지", "안심하지마라", "정신 똑디", "말아먹",
    "답답해뒤지겠다", "걱정이 더 크다", "번트 좀", "비맞게 놔두냐",
    "다칠까봐", "왜그러는지", "방출", "200구", "불안", "아마추어",
    "치워라", "보기 불편", "나락", "대신할", "왜이리", "챙겨줘야", "정신 차려라", "없나",
    "아님", "무섭다", "불편", "죠짓네"
]

# 감정 수정 함수 (직접 label 값 변경)
def apply_keyword_override(row):
    text = row["cleaned_comment"]
    original = row["label"]
    pos_flag = any(word in text for word in positive_keywords)
    neg_flag = any(word in text for word in negative_keywords)

    if original == "negative" and pos_flag and not neg_flag:
        return "positive"
    if original == "positive" and neg_flag and not pos_flag:
        return "negative"
    return original

df_lotte["label_before"] = df_lotte["label"]
df_lotte["label"] = df_lotte.apply(apply_keyword_override, axis=1)
df_changed = df_lotte[df_lotte["label"] != df_lotte["label_before"]]

# 수정된 댓글 추출
print(f"감정이 수정된 댓글 수: {len(df_changed)}")
print(df_changed[["cleaned_comment", "label_before", "label"]].head(10))


# 6. 감정 비율 계산
sentiment_counts = df_lotte["label"].value_counts(normalize=True) * 100

# 7. 감정 분포 시각화 (수정된 감정 기준)
colors = {"positive": "skyblue", "negative": "red"}
pie_colors = [colors[label] for label in sentiment_counts.index]

plt.figure(figsize=(6, 6))
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct="%1.1f%%",
    colors=pie_colors,
    startangle=140
)
plt.title("수정된 감정 분포 (긍정 vs 부정)", fontsize=14)
plt.tight_layout()
plt.show()
```
