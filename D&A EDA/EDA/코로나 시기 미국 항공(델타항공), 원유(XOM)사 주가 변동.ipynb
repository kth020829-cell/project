```python
#세계 3대 석유종의 변화 추이가 비슷함을 나타낸 산점도
plt.scatter(oil['Dubai'], oil['Brent'])
```

```python
#세계 3대 석유종의 변화 추이가 비슷함을 나타낸 상광계수 (1로 갈수록 선형도가 높음) -> 결국 3개 중 아무 데이터나 써도 괜찮음
oil['Brent'].corr(oil['WTI'])
```

```python
# 아래부터는 원유가격변동과 원유사 주가변동의 선형관계를 알아봄 
```

```python
#선형도를 나타내기위해 원유가격변동 데이터에 맞게 20,21년도 원유사 주가 데이터 추출
nxom = xom[::-1].reset_index(drop=True)
nxom = nxom[12:36].reset_index(drop=True)
nxom
```

```python
#세계 3대 유종 중 두바이와 엑슨모빌사의 종가의 선형도는 양의 선형관계를 가짐
sns.regplot(x=oil['Dubai'], y=nxom['종가'])
```

```python
#상관계수가 0.7이 넘어감에 따라 양의 선형관계라는 주장을 뒷받쳐줌 -> 원유가 올라가거나 내려감에 따라 원유사(엑슨모빌)의 주가도 올라가거나 내려감
oil['Dubai'].corr(nxom['종가'])
```
