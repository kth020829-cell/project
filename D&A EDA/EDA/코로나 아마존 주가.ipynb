```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.font_manager as fm 
fm.fontManager.ttflist
[f.name for f in fm.fontManager.ttflist] 
```

```python
#데이터 불러오기
data = pd.read_csv(r"C:\Users\touch\OneDrive\바탕 화면\AMZN 과거 데이터.csv", encoding='utf-8')

# % 기호제거 및 숫자형으로 변환하는 함수
def remove(i):
    if isinstance(i, str):
        return float(i.rstrip('%'))
    return i

# '변동 %'컬럼에 함수 적용
data['변동 %'] = data['변동 %'].apply(remove)

#변수저장
x = data['날짜'].values
y = data['종가'].values
z = data['변동 %'].values

#변동치 바차트로 시각화
plt.rcParams['font.family'] = 'Gulim'

plt.bar(x[::-1], z[::-1])

plt.title('19-22년도 아마존 주가 월별 변동 %')
plt.xlabel('19-22년(1~12월)')
plt.ylabel('변동치')
x_ticks = np.arange(0, 48, 12) 
x_labels = [2019,2020,2021,2022] 
plt.xticks(x_ticks, x_labels) 
```
