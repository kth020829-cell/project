```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm 
fm.fontManager.ttflist
[f.name for f in fm.fontManager.ttflist] 
```

```python
data = pd.read_csv(r"C:\Users\touch\OneDrive\바탕 화면\경제활동별_GDP_및_GNI_계절조정__명목__분기__20240514105928.csv", encoding='utf-8')
data
```

```python
df = data[data['계정항목별']=='국내총생산(시장가격, GDP)']
df = df.set_index('계정항목별')
df
```

```python
#x축
x = ['1분기', '2분기', '3분기', '4분기']

#y축
c1 = df.iloc[:,:4]
c2 = df.iloc[:,4:8]
c3 = df.iloc[:,8:12]
c4 = df.iloc[:,12:16]
c5 = df.iloc[:,16:20]
c6 = df.iloc[:,20:24]

# c1.values[0]으로 1차원 배열 추출
```

```python
plt.rcParams['font.family'] = 'NanumGothic'

plt.plot(x,c1.values[0], marker = 'o', label = '2018')
plt.plot(x,c2.values[0], marker = 'o', label = '2019')
plt.plot(x,c3.values[0], marker = 'o', label = '2020')
plt.plot(x,c4.values[0], marker = 'o', label = '2021')
plt.plot(x,c5.values[0], marker = 'o', label = '2022')
plt.plot(x,c6.values[0], marker = 'o', label = '2023')

plt.title('국내총생산(GDP)',fontdict={'size':20})
plt.xlabel('분기', loc = 'right')
plt.ylabel('분기당 생산치', loc = 'top')
plt.legend(loc = 'upper right')
```
